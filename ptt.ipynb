```python
#融合一起

import urllib.request as req
import bs4 as bs
import json
import os


def get_meta(url):
# url = "https://www.ptt.cc/bbs/Beauty/M.1745554518.A.3E7.html"
  r = req.Request(url)
  r.add_header("user-agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36")
  response = req.urlopen(r)
  html = bs.BeautifulSoup(response.read())

  meta = html.find_all("span",{"class":"article-meta-value"})
  userid = meta[0].text
  board_name = meta[1].text
  postid = meta[2].text
  post_time = meta[3].text

  push_list = []
  pushes = html.find_all("div",{"class":"push"})

  for push in pushes:
    push_meta = push.find_all("span") #
    p_type = push_meta[0].text
    if "推"in p_type:
      push_type = 1
    elif "噓"in p_type:
      push_type = -1
    else:
      push_type = 0
    p_id = push_meta[1].text
    p_content = push_meta[2].text.replace(":","")
    p_ipandtime = push_meta[3].text.strip()
    p_dic = {
        "ID":p_id,
        "Content":p_content,
        "IP/date":p_ipandtime
    }
    push_list.append(p_dic)

  row = {
      "作者":userid,
      "看板":board_name,
      "標題":postid,
      "時間":post_time,
      "推文":push_list
  }
  fn = url.split("/")[-1].replace(".html",".json")
  with open(fn,"w",encoding="utf-8") as f:
    json.dump(row, f, ensure_ascii=False, indent=4)

def get_img(url):
  r = req.Request(url)
  r.add_header("user-agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36")
  response = req.urlopen(r)
  html = bs.BeautifulSoup(response.read())

  # 開啟資料夾
  dn = url.split("/")[-1]
  # 如果資料夾不存在, 我們就創造起來
  if not os.path.exists(dn):
    os.makedirs(dn)

  allow_sub = ["jpg","jpeg","png","gif"]
  links_list = html.find_all("a")
  for link in links_list:
    link = link["href"]
    sub = link.split(".")[-1]
    if sub.lower() in allow_sub:
      img_url = link
      # 加上資料夾名字
      fn = img_url.split("/")[-1]
      fn = "{}/{}".format(dn,fn)
      # save img
      r = req.Request(img_url)
      r.add_header("user-agent","Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36")
      img = req.urlopen(r)
      with open(fn,"wb") as f:
        f.write(img.read())

# url = "https://www.ptt.cc/bbs/Beauty/M.1745551662.A.900.html"
# get_meta(url)
# get_img(url)
```

```python
#將meta檔案轉換成表格
import json
import glob
import pandas as pd
table = []
for fn in glob.glob("*.json"): # *是萬用字元（wildcard），意思是「任意字串」
  with open(fn,"r",encoding="utf-8") as f:
    content = f.read()
  post = json.loads(content)
  table.append(post)
df = pd.DataFrame(table) #將table轉換為表格
df
```

```python
# 1. filter(過濾): 帶入跟你資料筆數依樣多的True/False
# 2. apply(轉換): apply(轉換函式)
def trans(s):
  return s.split("(")[0].strip()

df["pttid"] = df["作者"].apply(trans)
df
```
