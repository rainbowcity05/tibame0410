````python
#老師引導
import random
# 凱薩密碼的加密
s = "AppleZ"
# 以前: lower(s)
# 專屬功能: s.lower()
s_lower = s.lower()
# 位移幾個
password = random.randint(1, 25)

result = ""
for c in s_lower:
    # 取得ASCII->位移到0-25
    n = ord(c) - ord("a")
    # 往後位移做出密碼
    n = (n + password) % 26 + ord("a")
    # 放到result
    result = result + chr(n)
print(password, result)
````

````python
#第二題學生解法1:已知最後一段對應的文字

import urllib.request as req

url = "https://github.com/Elwing-Chou/tibame1224/raw/refs/heads/main/password.txt"
req.urlretrieve(url, "password.txt")
f = open("password.txt", "r", encoding="utf-8")
message = f.read()
f.close()

#找出密碼
password = (ord("x")-ord("i")) % 26

#位移文章
alphabet = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
m_lower = message.lower()
new_message = ""

for c in m_lower:
  if c in alphabet:
    n = (ord(c) - password - ord('a')) % 26 + ord('a')  #解密後的字母ord
    new_alphabet = chr(n) #轉換回字母
    new_message = new_message + new_alphabet
  else:
    new_message = new_message + c #標點符號或空格不做轉換

print(new_message)
````

````python
#第二題學生解法2:只知道原文中包含的字，但不知道對應位置

import urllib.request as req

url = "https://github.com/Elwing-Chou/tibame1224/raw/refs/heads/main/password.txt"
req.urlretrieve(url, "password.txt")
f = open("password.txt", "r", encoding="utf-8")
message = f.read()
f.close()


alphabet = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
m_lower = message.lower()

for i in range(1,26): #暴力破解法，從1試到25
  new_message = "" #每猜1次都清空字串
  for c in m_lower:
    if c in alphabet:
      n = (ord(c) - i - ord('a')) % 26 + ord('a')  #解密後的字母ord
      new_alphabet = chr(n) #轉換回字母
      new_message = new_message + new_alphabet
    else:
      new_message = new_message + c #標點符號或空格不做轉換
  if "trouble" in new_message:
    print("密碼是:",i,"\n",new_message)
  else:
    pass

````

````python
#老師引導
import ColabTurtle.Turtle as turtle
import math

turtle.initializeTurtle()
# 設置狀態
turtle.speed(10)
turtle.pensize(2)

mid_angle = 1
other_angle = (180 - mid_angle) / 2.0
other_angle_radians = math.radians(other_angle)

times = 360 // mid_angle

width = 100
other_width = 2 * 100 * math.cos(other_angle_radians)
for i in range(times):
    # penup: 不會畫
    turtle.penup()
    turtle.forward(width)
    turtle.right(180 - other_angle)
    # pendown: 才畫
    turtle.pendown()
    turtle.forward(other_width)
    turtle.penup()
    turtle.right(180 - other_angle)
    turtle.forward(width)
    turtle.right(180)
````

````python
#第三題學生解法

import ColabTurtle.Turtle as turtle
turtle.initializeTurtle()
# 設置狀態
turtle.speed(2)
turtle.pensize(2)

points = [{"x":200, "y":50},
          {"x":600, "y":50},
          {"x":600, "y":450},
          {"x":200, "y":450}]

# 這個東西應該在一個迴圈裡
turtle.penup()
x, y = points[0]["x"], points[0]["y"]
# 把烏龜移到開頭
turtle.goto(x, y)
turtle.pendown()

# 把四個點連起來
for i in range(4):
    # goto next
    x, y = points[(i+1)%4]["x"], points[(i+1)%4]["y"]
    turtle.goto(x, y)

total_layers = 9 #層數

current_points = points  # 用上一層的四組點當成基準座標

for l in range(total_layers):
    new_points = [] # 計算新的4組中間點
    for i in range(4):
        x1, y1 = current_points[i]["x"], current_points[i]["y"]
        x2, y2 = current_points[(i+1)%4]["x"], current_points[(i+1)%4]["y"]
        mid_x = (x1 + x2) / 2 #新的點座標是上一組點的邊長的一半
        mid_y = (y1 + y2) / 2
        new_points.append({"x": mid_x, "y": mid_y}) #根據上一層的四組座標做出新的四組座標

    # 畫出這一層
    turtle.penup()
    turtle.goto(new_points[0]["x"], new_points[0]["y"])
    turtle.pendown()
    for i in range(4):
        x, y = new_points[(i+1)%4]["x"], new_points[(i+1)%4]["y"]
        turtle.goto(x, y)

    # 更新current_points的基準座標
    current_points = new_points
````
