````python
import json
import glob
import pandas as pd
table = []
for fn in glob.glob("*.json"):
    with open(fn, "r", encoding="utf-8") as f:
        content = f.read()
    post = json.loads(content)
    table.append(post)
df = pd.json_normalize(table)
df
````

````python
# 1. filter(過濾): 帶入跟你資料筆數依樣多的True/False
# 2. apply(轉換): apply(轉換函式)
def trans(s):
    # s = "willy911006 (小溫)	"
    return s.split("(")[0].strip()

df["pttid"] = df["作者"].apply(trans)
df
````

````python
import datetime as dt
def trans(s):
    # dt_s = "Wed Apr 23 11:49:57 2025"
    dt_s = s.strip()
    dt_parse = dt.datetime.strptime(dt_s, "%a %b %d %H:%M:%S %Y")
    dt_s_trans = dt_parse.strftime("%Y/%m/%d %H:%M:%S")
    return dt_s_trans

df["時間"].apply(trans)
````
